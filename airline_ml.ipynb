```python
import numpy as np
import pandas as pd
import pickle
import time
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


import warnings
warnings.filterwarnings('ignore')

%config InlineBackend.figure_formats = ['svg']
```

```python
airline_df = pd.read_csv('airline.csv')
airline_df.head()
```

```python
airline_df.dropna(inplace=True)
airline_df.info()
```

```python
output = airline_df['satisfaction']
input = airline_df.drop(columns = ['satisfaction'])
input.head()
```

```python
dum_input = pd.get_dummies(input)
dum_input.head()
```

```python
train_X, test_X, train_y, test_y = train_test_split(dum_input, output, test_size = 0.2, random_state = 1)
```

```python
tree_class = DecisionTreeClassifier(random_state = 1)
```

```python
start = time.time()
tree_class.fit(train_X, train_y)
stop = time.time()
print(f'Time Taken: {stop - start} seconds')
```

```python
y_pred_train = tree_class.predict(train_X)
```

```python
y_pred = tree_class.predict(test_X)
```

```python
probs = tree_class.predict_proba(test_X)
probs
```

```python
dt_class = open('airline_dt.pickle', 'wb')
pickle.dump(tree_class, dt_class)
dt_class.close()
```
